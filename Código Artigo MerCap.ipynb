# CAPM

$r_i-r_f = \beta_1\cdot(r_m-r_f) + \epsilon$

# Modelo de 3 fatores:

$r_i-r_f = \beta_1\cdot(r_m-r_f) +\beta_2\cdot (SMB) + \beta_3\cdot (HML) + \epsilon$

# Modelo de 5 fatores: 

$r_i-r_f = \beta_1\cdot(r_m-r_f) +\beta_2\cdot (SMB) + \beta_3\cdot (HML) + \beta_4\cdot (IML) +\beta_5\cdot (WML) + \epsilon$
